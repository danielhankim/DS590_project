```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
```

```python
walk_df = pd.read_csv('../data/processed/Bloomington_CBG_Amenity_15_Minute_Walk.csv')
bike_df = pd.read_csv('../data/processed/Bloomington_CBG_Amenity_15_Minute_Bike.csv')
```

```python
walk_df.drop(['Unnamed: 0', 'TRACTCE', 'rep_x', 'rep_y', 'node_id',], axis=1, inplace=True)
walk_df.sample(5)
```

```python
bike_df.drop(['Unnamed: 0', 'TRACTCE', 'rep_x', 'rep_y', 'node_id'], axis=1, inplace=True)
bike_df.sample(5)
```

```python
walk_df['mode'] = 'walk'
bike_df['mode'] = 'bike'

df = pd.concat([walk_df, bike_df])
df
```

```python
time_df = df[['sustenance_time', 'education_time', 'financial_time', 
              'healthcare_time', 'culture_time', 'public_time', 'mode']]

time_df
```

```python
g = sns.pairplot(time_df, hue='mode')

for ax in g.axes.ravel():
    ax.axvline(x=15, ls='--', c='k')
plt.show()
```
